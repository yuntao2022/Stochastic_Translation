# Unified Binomial Moment Analysis

This notebook consolidates four binomial-moment examples into one reusable workflow.

The shared model, numerical routines, result export, visualization, and optional probability-mass-function reconstruction are defined once. Each example is represented only by its case-specific parameters.

A geometric recurrence is used to avoid the repeated quadratic convolution in the original notebooks. For
\(C_r = q^{r-1} C_1\), the convolution satisfies

\[
S_m = Z_{m-1} C_1 + q S_{m-1},
\]

which reduces the moment calculation from \(O(M_{\max}^2)\) matrix operations to \(O(M_{\max})\).

```python
from __future__ import annotations

from dataclasses import dataclass
from numbers import Integral
from pathlib import Path
from typing import Optional
import math

import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import solve


np.set_printoptions(precision=8, suppress=True)

# Global execution options.
OUTPUT_DIR = Path("binomial_results")
SAVE_RESULTS = True
PLOT_RESULTS = True
PRINT_FULL_SEQUENCES = False

# PMF reconstruction is optional because long alternating sums may be
# sensitive to floating-point cancellation for large truncation orders.
RECONSTRUCT_PMF = False
```

## Model and case configuration

The four examples use the same matrices and differ only in the geometric parameter, truncation order, plotting range, and output index.

```python
@dataclass(frozen=True)
class CaseConfig:
    index: int
    name: str
    lam: float
    m_max: int
    plot_points: int
    n_max: int = 31
    delta: float = 1.0

    @property
    def moment_filename(self) -> str:
        return f"BM_real{self.index}.npy"

    @property
    def bound_filename(self) -> str:
        return f"BM_bound{self.index}.npy"

    @property
    def pmf_filename(self) -> str:
        return f"PMF{self.index}.npy"


D0 = np.zeros((4, 4), dtype=float)

D1 = np.array(
    [
        [0.2, 0.2, 0.1, 0.8],
        [0.2, 0.2, 0.1, 0.8],
        [0.1, 0.1, 0.1, 0.0],
        [0.8, 0.8, 0.0, 0.8],
    ],
    dtype=float,
)

CASES = (
    CaseConfig(index=1, name="Case 1", lam=0.20, m_max=1000, plot_points=1000),
    CaseConfig(index=2, name="Case 2", lam=0.49, m_max=1100, plot_points=1000),
    CaseConfig(index=3, name="Case 3", lam=0.51, m_max=1010, plot_points=1000),
    CaseConfig(index=4, name="Case 4", lam=0.60, m_max=510, plot_points=500),
)
```

## Core numerical routines

The functions below validate the model, construct the infinitesimal generator, solve for the stationary distribution, calculate binomial moments and bounds, and optionally reconstruct a truncated probability mass function.

```python
@dataclass
class CaseResult:
    config: CaseConfig
    generator: np.ndarray
    stationary: np.ndarray
    moments: np.ndarray
    bounds: np.ndarray
    stationary_residual: float
    probabilities: Optional[np.ndarray] = None


def validate_model(d0: np.ndarray, d1: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """Validate and convert the two model matrices."""
    d0 = np.asarray(d0, dtype=float)
    d1 = np.asarray(d1, dtype=float)

    if d0.shape != d1.shape:
        raise ValueError("D0 and D1 must have identical shapes.")
    if d0.ndim != 2 or d0.shape[0] != d0.shape[1]:
        raise ValueError("D0 and D1 must be square matrices.")
    if not np.all(np.isfinite(d0)) or not np.all(np.isfinite(d1)):
        raise ValueError("D0 and D1 must contain only finite values.")

    return d0, d1


def validate_case(config: CaseConfig) -> None:
    """Validate one parameter set before running the calculation."""
    if not 0.0 < config.lam < 1.0:
        raise ValueError("lam must satisfy 0 < lam < 1.")
    if config.m_max < 0:
        raise ValueError("m_max must be non-negative.")
    if config.n_max < 1:
        raise ValueError("n_max must be at least 1.")
    if config.plot_points < 1:
        raise ValueError("plot_points must be at least 1.")
    if config.delta <= 0.0:
        raise ValueError("delta must be positive.")


def build_geometric_generator(
    d0: np.ndarray,
    d1: np.ndarray,
    lam: float,
) -> np.ndarray:
    """Construct the generator and enforce zero row sums."""
    generator = np.asarray(d0, dtype=float) + np.asarray(d1, dtype=float) / (1.0 - lam)
    generator = generator.copy()

    # Replace each diagonal entry with the negative sum of its off-diagonal row.
    off_diagonal_sums = generator.sum(axis=1) - np.diag(generator)
    np.fill_diagonal(generator, -off_diagonal_sums)
    return generator


def stationary_distribution(generator: np.ndarray) -> np.ndarray:
    """Solve pi D = 0 together with sum(pi) = 1."""
    n_states = generator.shape[0]
    system = generator.T.copy()
    system[0, :] = 1.0

    rhs = np.zeros(n_states, dtype=float)
    rhs[0] = 1.0

    pi = solve(system, rhs, assume_a="gen", check_finite=False)
    pi /= pi.sum()
    return pi


def compute_binomial_moments(
    generator: np.ndarray,
    d1: np.ndarray,
    lam: float,
    pi: np.ndarray,
    m_max: int,
    delta: float = 1.0,
) -> np.ndarray:
    """Compute moments B_0 through B_m_max using the geometric recurrence.

    The original convolution is
        S_m = sum_{i=1}^m Z_{m-i} C_i.

    Since C_i = q^(i-1) C_1 for q = lam / (1 - lam),
        S_m = Z_{m-1} C_1 + q S_{m-1}.

    This avoids storing every C_i and every Z_i.
    """
    n_states = generator.shape[0]
    identity = np.eye(n_states)
    ones = np.ones((n_states, 1))
    pi_row = np.asarray(pi, dtype=float).reshape(1, -1)

    c1 = np.asarray(d1, dtype=float) / (1.0 - lam) ** 2
    ratio = lam / (1.0 - lam)

    z_previous = identity
    convolution_previous = np.zeros_like(generator, dtype=float)

    moments = np.empty(m_max + 1, dtype=float)
    moments[0] = 1.0

    for order in range(1, m_max + 1):
        convolution = z_previous @ c1 + ratio * convolution_previous
        coefficient = order * delta * identity - generator

        # Solve Z_m * coefficient = convolution without forming an inverse.
        z_current = solve(
            coefficient.T,
            convolution.T,
            assume_a="gen",
            check_finite=False,
        ).T

        moments[order] = (pi_row @ z_current @ ones).item()
        z_previous = z_current
        convolution_previous = convolution

    return moments


def compute_binomial_bounds(
    m_max: int,
    lam: float,
    d1_infinity_norm: float,
    delta: float = 1.0,
) -> np.ndarray:
    """Compute bounds a_0 through a_m_max with matching array length."""
    alpha = d1_infinity_norm / (delta * lam * (1.0 - lam))
    ratio = lam / (1.0 - lam)

    bounds = np.empty(m_max + 1, dtype=float)
    bounds[0] = 1.0

    with np.errstate(over="ignore", under="ignore", invalid="ignore"):
        for order in range(m_max):
            bounds[order + 1] = (
                bounds[order]
                * ratio
                * (alpha + order)
                / (order + 1)
            )

    return bounds


def stable_scaled_combination(n: int, k: int, scale: float) -> float:
    """Return binomial(n, k) * scale using logarithmic gamma functions."""
    if not isinstance(n, Integral) or not isinstance(k, Integral):
        raise TypeError("n and k must be integers.")
    if n < 0:
        raise ValueError("n must be non-negative.")
    if k < 0 or k > n:
        return 0.0

    scale = float(scale)
    if scale < 0.0:
        raise ValueError("scale must be non-negative.")
    if scale == 0.0:
        return 0.0

    log_combination = (
        math.lgamma(n + 1)
        - math.lgamma(k + 1)
        - math.lgamma(n - k + 1)
    )
    log_value = log_combination + math.log(scale)

    log_tiny = math.log(np.finfo(float).tiny)
    log_max = math.log(np.finfo(float).max)

    if log_value < log_tiny:
        return 0.0
    if log_value > log_max:
        return math.inf

    return math.exp(log_value)


def reconstruct_probability_mass_function(
    moments: np.ndarray,
    n_max: int,
) -> np.ndarray:
    """Reconstruct probabilities for n = 0, ..., n_max - 1.

    The calculation uses the truncated alternating series
        p_n = sum_{j=n}^M (-1)^(j-n) binomial(j, n) B_j.
    """
    moments = np.asarray(moments, dtype=float)
    probabilities = np.empty(n_max, dtype=float)

    for n in range(n_max):
        terms = []
        for j in range(n, moments.size):
            magnitude = stable_scaled_combination(j, n, moments[j])
            terms.append(magnitude if (j - n) % 2 == 0 else -magnitude)
        probabilities[n] = math.fsum(terms)

    return probabilities


def run_case(
    config: CaseConfig,
    d0: np.ndarray,
    d1: np.ndarray,
    reconstruct_pmf: bool = False,
) -> CaseResult:
    """Run the complete calculation for one case."""
    validate_case(config)
    d0, d1 = validate_model(d0, d1)

    generator = build_geometric_generator(d0, d1, config.lam)
    pi = stationary_distribution(generator)

    moments = compute_binomial_moments(
        generator=generator,
        d1=d1,
        lam=config.lam,
        pi=pi,
        m_max=config.m_max,
        delta=config.delta,
    )

    bounds = compute_binomial_bounds(
        m_max=config.m_max,
        lam=config.lam,
        d1_infinity_norm=np.linalg.norm(d1, ord=np.inf),
        delta=config.delta,
    )

    residual = float(np.max(np.abs(pi @ generator)))

    probabilities = None
    if reconstruct_pmf:
        probabilities = reconstruct_probability_mass_function(
            moments=moments,
            n_max=config.n_max,
        )

    return CaseResult(
        config=config,
        generator=generator,
        stationary=pi,
        moments=moments,
        bounds=bounds,
        stationary_residual=residual,
        probabilities=probabilities,
    )


def print_numbered_sequence(label: str, values: np.ndarray) -> None:
    """Print every element using the format of the original notebooks."""
    for index, value in enumerate(values):
        print(f"{label}[{index:4d}] = {float(value):.6e}")
```

## Run all four cases

All shared calculations are executed through the same `run_case` function. The result dictionary is keyed by the original case index.

```python
results = {
    config.index: run_case(
        config=config,
        d0=D0,
        d1=D1,
        reconstruct_pmf=RECONSTRUCT_PMF,
    )
    for config in CASES
}

for result in results.values():
    config = result.config
    print(
        f"{config.name}: "
        f"lambda={config.lam:.2f}, "
        f"M_max={config.m_max}, "
        f"stationary={result.stationary}, "
        f"max|pi D|={result.stationary_residual:.3e}, "
        f"B_M={result.moments[-1]:.6e}, "
        f"bound_M={result.bounds[-1]:.6e}"
    )

    if PRINT_FULL_SEQUENCES:
        print_numbered_sequence("moment", result.moments)
        print_numbered_sequence("bound", result.bounds)
        if result.probabilities is not None:
            print_numbered_sequence("probability", result.probabilities)
        print()
```

## Save numerical results

The original output names are preserved and placed in one dedicated directory.

```python
def save_case_result(result: CaseResult, output_dir: Path) -> list[Path]:
    """Save moments, bounds, and optional PMF data for one case."""
    output_dir.mkdir(parents=True, exist_ok=True)

    moment_path = output_dir / result.config.moment_filename
    bound_path = output_dir / result.config.bound_filename

    np.save(moment_path, result.moments)
    np.save(bound_path, result.bounds)

    saved_paths = [moment_path, bound_path]

    if result.probabilities is not None:
        pmf_path = output_dir / result.config.pmf_filename
        np.save(pmf_path, result.probabilities)
        saved_paths.append(pmf_path)

    return saved_paths


if SAVE_RESULTS:
    saved_files = []
    for result in results.values():
        saved_files.extend(save_case_result(result, OUTPUT_DIR))

    print(f"Saved {len(saved_files)} files to: {OUTPUT_DIR.resolve()}")
    for path in saved_files:
        print(f"  {path}")
```

## Compare moments and bounds

Each case is displayed once on a linear scale and once on a logarithmic scale. Non-finite values and non-positive values on the logarithmic plot are excluded safely.

```python
def finite_values(values: np.ndarray) -> np.ndarray:
    """Convert non-finite values to NaN for safe plotting."""
    array = np.asarray(values, dtype=float).copy()
    array[~np.isfinite(array)] = np.nan
    return array


def plot_moments_and_bounds(result: CaseResult) -> None:
    """Create separate linear-scale and logarithmic-scale figures."""
    config = result.config
    point_count = min(
        config.plot_points,
        result.moments.size,
        result.bounds.size,
    )

    orders = np.arange(point_count)
    moments = finite_values(result.moments[:point_count])
    bounds = finite_values(result.bounds[:point_count])

    plt.figure(figsize=(8, 5))
    plt.plot(orders, moments, label="Binomial moments")
    plt.plot(orders, bounds, label="Binomial bounds")
    plt.xlabel("Moment order")
    plt.ylabel("Value")
    plt.title(f"{config.name}: moments and bounds, lambda={config.lam:.2f}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    positive_moments = np.where(moments > 0.0, moments, np.nan)
    positive_bounds = np.where(bounds > 0.0, bounds, np.nan)

    plt.figure(figsize=(8, 5))
    plt.semilogy(orders, positive_moments, label="Binomial moments")
    plt.semilogy(orders, positive_bounds, label="Binomial bounds")
    plt.xlabel("Moment order")
    plt.ylabel("Value")
    plt.title(f"{config.name}: logarithmic comparison, lambda={config.lam:.2f}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


if PLOT_RESULTS:
    for result in results.values():
        plot_moments_and_bounds(result)
```

## Optional probability mass function

Set `RECONSTRUCT_PMF = True` in the options cell and rerun the notebook to calculate and plot the truncated probability mass function for every case. Large alternating sums can suffer from cancellation, so the result should be checked for normalization and small negative numerical artifacts.

```python
def plot_probability_mass_function(result: CaseResult) -> None:
    """Plot the reconstructed PMF when it is available."""
    if result.probabilities is None:
        return

    indices = np.arange(result.probabilities.size)

    print(
        f"{result.config.name}: "
        f"PMF sum={result.probabilities.sum():.12g}, "
        f"minimum={result.probabilities.min():.6e}"
    )

    plt.figure(figsize=(8, 5))
    plt.stem(indices, result.probabilities)
    plt.xlabel("mRNA copy number")
    plt.ylabel("Probability")
    plt.title(f"{result.config.name}: reconstructed probability mass function")
    plt.grid(True)
    plt.tight_layout()
    plt.show()


if RECONSTRUCT_PMF:
    for result in results.values():
        plot_probability_mass_function(result)
else:
    print("PMF reconstruction is disabled. Set RECONSTRUCT_PMF = True to enable it.")
```
