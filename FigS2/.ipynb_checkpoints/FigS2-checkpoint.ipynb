```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter

# 全局样式
plt.rcParams['axes.linewidth'] = 2

# 最多显示前多少个点
x_max_list = [51, 1001, 1001, 501]
lam=[0.2,0.49,0.51,0.6]

# 创建 2x2 子图
fig, axes = plt.subplots(2, 2, figsize=(14, 14), dpi=100)
axes = axes.flatten()

for i in range(1, 5):
    ax = axes[i - 1]
    x_max = x_max_list[i - 1]

    # 读取第 i 组数据
    BM_real = np.load(f'BM_real{i}.npy', allow_pickle=True).flatten()
    BM_bound = np.load(f'BM_bound{i}.npy', allow_pickle=True).flatten()

    # 转成 float
    BM_real = np.array(BM_real, dtype=float)
    BM_bound = np.array(BM_bound, dtype=float)

    # 统一取前 x_max 个，并保证长度一致
    n = min(x_max, len(BM_real), len(BM_bound))
    x = np.arange(n)

    real_plot = BM_real[:n]
    bound_plot = BM_bound[:n]

    # 绘制两条曲线
    ax.plot(
        x,
        real_plot,
        linestyle='--',
        marker='o',
        markersize=8,
        linewidth=1,
        alpha=0.8,
        label='Binomial Moments'
    )

    ax.plot(
        x,
        bound_plot,
        linestyle='-',
        marker='v',
        markersize=8,
        linewidth=1,
        alpha=0.8,
        label='Upper Bounds'
    )

    # 横轴范围
    ax.set_xlim(0, n - 1)

    # 前两张图：线性坐标；后两张图：log 坐标
    if i <= 2:
        y_max = np.nanmax(real_plot)
        ax.set_ylim(0, 1.05 * y_max if y_max > 0 else 1)
    else:
        ax.set_yscale('log')

        # log 轴下界必须是正数
        positive_vals = np.concatenate([
            real_plot[np.isfinite(real_plot) & (real_plot > 0)],
            bound_plot[np.isfinite(bound_plot) & (bound_plot > 0)]
        ])

        if len(positive_vals) > 0:
            y_min = np.min(positive_vals)
            y_max = np.max(positive_vals)
            ax.set_ylim(y_min, 1.05 * y_max)
        else:
            ax.set_ylim(1e-12, 1)

    # 子图标题
    ax.set_title(
        rf'Example {i} ($\lambda={lam[i-1]:.2f}$)',
        fontname='Times New Roman',
        fontsize=25,
        fontweight='bold',
        color='green'
    )

    row = (i - 1) // 2
    col = (i - 1) % 2

    # 只有最左边显示 y 轴标题
    if col == 0:
        ax.set_ylabel(
            'Value',
            fontsize=20,
            fontweight='bold',
            fontname='Times New Roman',
            color='green'
        )

    # 只有最下面显示 x 轴标题
    if row == 1:
        ax.set_xlabel(
            'Order of Binomial Moments',
            fontsize=20,
            fontweight='bold',
            fontname='Times New Roman',
            color='green'
        )

    # 刻度样式
    ax.tick_params(direction='out', width=2, length=6, labelsize=12)

    # 第二幅图：调整 y 轴顶部的 1e7
    if i == 2:
        formatter = ScalarFormatter(useMathText=True)
        formatter.set_scientific(True)
        formatter.set_powerlimits((0, 0))
        ax.yaxis.set_major_formatter(formatter)

        offset_text = ax.yaxis.get_offset_text()
        offset_text.set_fontname('Times New Roman')
        offset_text.set_fontsize(14)
        offset_text.set_color('blue')
        offset_text.set_x(-0.02)   # 左右调节
        offset_text.set_y(1.01)    # 上下调节

    for label in ax.get_xticklabels():
        label.set_fontname('Times New Roman')
        label.set_color('blue')

    for label in ax.get_yticklabels():
        label.set_fontname('Times New Roman')
        label.set_color('blue')
        label.set_rotation(90)

    # 网格
    ax.grid(True, linestyle='--', alpha=0.3)

    # 图例
    legend = ax.legend(
        loc='upper right',
        frameon=True,
        shadow=True,
        markerscale=1.2,
        fancybox=True,
        prop={'family': 'Times New Roman', 'size': 20}
    )
    legend.get_frame().set_color('#E6E6FA')
    legend.get_frame().set_alpha(0.5)

    # 去掉边框
    for spine in ax.spines.values():
        spine.set_visible(False)

# 总标题
fig.suptitle(
    'Binomial Moments and Upper Bounds',
    fontname='Times New Roman',
    fontsize=30,
    fontweight='bold',
    color='green'
)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('BM_2x2.pdf', format='pdf', bbox_inches='tight', dpi=600)
plt.show()
```
