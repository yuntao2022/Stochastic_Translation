# Analytical Results based on Binomial Moments

```python
from __future__ import annotations

from dataclasses import dataclass
from pathlib import Path
from time import perf_counter
from typing import Dict, Iterable, Literal, Optional

import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import inv, solve
```

```python
@dataclass(frozen=True)
class ExampleConfig:
    """Parameters for one numerical example."""

    example_id: int
    name: str
    D0: np.ndarray
    D1: np.ndarray
    lam: float
    delta: float
    M_max: int
    N_max: int
    r_max: int
    output_prefix: str

    @property
    def N(self) -> int:
        """Return the number of model states."""
        return int(self.D0.shape[0])


@dataclass
class ExampleResult:
    """Outputs from one complete model run."""

    config: ExampleConfig
    D: np.ndarray
    pi: np.ndarray
    binomial_moments: np.ndarray
    pmf: np.ndarray
    probability_sum: float
    elapsed_seconds: float
    method: str


def validate_config(config: ExampleConfig) -> None:
    """Validate matrix shapes and numerical parameters."""
    D0 = np.asarray(config.D0)
    D1 = np.asarray(config.D1)

    if D0.ndim != 2 or D0.shape[0] != D0.shape[1]:
        raise ValueError("D0 must be a square matrix.")

    if D1.shape != D0.shape:
        raise ValueError("D1 must have the same shape as D0.")

    if not 0.0 <= config.lam < 1.0:
        raise ValueError("lam must satisfy 0 <= lam < 1.")

    if config.delta <= 0:
        raise ValueError("delta must be positive.")

    if config.M_max < 0:
        raise ValueError("M_max must be non-negative.")

    if config.N_max <= 0:
        raise ValueError("N_max must be positive.")

    if config.r_max < config.M_max:
        raise ValueError("r_max must be at least M_max for the direct method.")

    if not np.all(np.isfinite(D0)) or not np.all(np.isfinite(D1)):
        raise ValueError("D0 and D1 must contain only finite values.")


def geometric_coefficient_matrices(
    D1: np.ndarray,
    lam: float,
    r_max: int,
) -> list[Optional[np.ndarray]]:
    """Build one-based coefficient matrices for the direct recursion."""
    D1 = np.asarray(D1, dtype=float)

    # Index 0 is unused so coefficients[r] matches C_r.
    coefficients: list[Optional[np.ndarray]] = [None]

    factor = 1.0 / (1.0 - lam) ** 2

    ratio = lam / (1.0 - lam)

    for _ in range(1, r_max + 1):
        coefficients.append(factor * D1)
        factor *= ratio

    return coefficients


def effective_generator(
    D0: np.ndarray,
    D1: np.ndarray,
    lam: float,
) -> np.ndarray:
    """Build the generator and enforce zero row sums."""
    D0 = np.asarray(D0, dtype=float)
    D1 = np.asarray(D1, dtype=float)

    D = D0 + D1 / (1.0 - lam)
    D = D.copy()

    off_diagonal = D.copy()
    np.fill_diagonal(off_diagonal, 0.0)

    # Enforce zero row sums.
    np.fill_diagonal(D, -off_diagonal.sum(axis=1))

    return D


def stationary_distribution(D: np.ndarray) -> np.ndarray:
    """Solve the stationary distribution with normalization."""
    D = np.asarray(D, dtype=float)
    N = D.shape[0]

    system = D.T.copy()

    # Replace one stationary equation with normalization.
    system[0, :] = 1.0

    rhs = np.zeros(N)
    rhs[0] = 1.0

    pi = np.linalg.solve(system, rhs)
    return pi


def binomial_moments_direct(
    D: np.ndarray,
    C_list: list[Optional[np.ndarray]],
    pi: np.ndarray,
    M_max: int,
    delta: float = 1.0,
) -> np.ndarray:
    """Compute binomial moments with the original quadratic recursion."""
    pi_row = np.asarray(pi, dtype=float).reshape(1, -1)

    N = D.shape[0]
    ones = np.ones((N, 1))
    identity = np.eye(N)

    # Z_0 = I; the zeroth moment is 1.
    Z_list = [identity]

    moments = np.empty(M_max + 1, dtype=float)
    moments[0] = 1.0

    for m in range(1, M_max + 1):
        S = np.zeros((N, N))

        # S_m = sum Z_(m-i) C_i.
        for i in range(1, m + 1):
            C_i = C_list[i]

            if C_i is None:
                raise RuntimeError(f"C_list[{i}] is missing.")

            S += Z_list[m - i] @ C_i

        matrix = m * delta * identity - D

        Z_m = S @ inv(matrix, overwrite_a=False)

        Z_list.append(Z_m)

        moments[m] = float((pi_row @ Z_m @ ones)[0, 0])

    return moments


def binomial_moments_recursive(
    D: np.ndarray,
    D1: np.ndarray,
    lam: float,
    pi: np.ndarray,
    M_max: int,
    delta: float = 1.0,
) -> np.ndarray:
    """Compute moments with the linear-cost geometric recurrence."""
    pi_row = np.asarray(pi, dtype=float).reshape(1, -1)
    D1 = np.asarray(D1, dtype=float)

    N = D.shape[0]
    ones = np.ones((N, 1))
    identity = np.eye(N)

    moments = np.empty(M_max + 1, dtype=float)
    moments[0] = 1.0

    Z_previous = identity

    convolution_previous = np.zeros((N, N))

    first_factor = 1.0 / (1.0 - lam) ** 2
    ratio = lam / (1.0 - lam)

    for m in range(1, M_max + 1):
        # Update the geometric convolution recursively.
        convolution = (
            first_factor * Z_previous
            + ratio * convolution_previous
        )

        S = convolution @ D1

        matrix = m * delta * identity - D

        # Solve the transposed system instead of forming an inverse.
        Z_m = solve(
            matrix.T,
            S.T,
            assume_a="gen",
            check_finite=False,
        ).T

        moments[m] = float((pi_row @ Z_m @ ones)[0, 0])

        Z_previous = Z_m
        convolution_previous = convolution

    return moments


def scaled_combination(n: int, k: int, scale: float) -> float:
    """Compute comb(n, k) times scale in the log domain."""
    if not isinstance(n, (int, np.integer)):
        raise TypeError("n must be an integer.")

    if not isinstance(k, (int, np.integer)):
        raise TypeError("k must be an integer.")

    if n < 0 or k < 0 or k > n:
        raise ValueError("Require n >= 0 and 0 <= k <= n.")

    scale = float(scale)

    if scale < 0:
        raise ValueError(f"scale must be non-negative; received {scale}.")

    if scale == 0:
        return 0.0

    log_value = (
        math.lgamma(n + 1)
        - math.lgamma(k + 1)
        - math.lgamma(n - k + 1)
        + math.log(scale)
    )

    # Avoid underflow warnings.
    if log_value < -700:
        return 0.0

    return math.exp(log_value)


def reconstruct_pmf(
    binomial_moments: np.ndarray,
    N_max: int,
) -> np.ndarray:
    """Recover the truncated PMF by binomial inversion."""
    moments = np.asarray(binomial_moments, dtype=float)
    M_max = len(moments) - 1

    pmf = np.zeros(N_max, dtype=float)

    for n in range(N_max):
        total = 0.0

        sign = 1.0

        for j in range(n, M_max + 1):
            total += sign * scaled_combination(j, n, moments[j])
            sign = -sign

        pmf[n] = total

    return pmf


def run_example(
    config: ExampleConfig,
    *,
    method: Literal["direct", "recursive"] = "direct",
    verbose: bool = True,
) -> ExampleResult:
    """Run one example from validation through PMF reconstruction."""
    validate_config(config)
    started = perf_counter()

    D = effective_generator(config.D0, config.D1, config.lam)

    pi = stationary_distribution(D)

    if method == "direct":
        C_list = geometric_coefficient_matrices(
            config.D1,
            config.lam,
            config.r_max,
        )

        moments = binomial_moments_direct(
            D=D,
            C_list=C_list,
            pi=pi,
            M_max=config.M_max,
            delta=config.delta,
        )

    elif method == "recursive":
        moments = binomial_moments_recursive(
            D=D,
            D1=config.D1,
            lam=config.lam,
            pi=pi,
            M_max=config.M_max,
            delta=config.delta,
        )

    else:
        raise ValueError("method must be 'direct' or 'recursive'.")

    pmf = reconstruct_pmf(
        binomial_moments=moments,
        N_max=config.N_max,
    )

    elapsed = perf_counter() - started

    result = ExampleResult(
        config=config,
        D=D,
        pi=pi,
        binomial_moments=moments,
        pmf=pmf,
        probability_sum=float(pmf.sum()),
        elapsed_seconds=elapsed,
        method=method,
    )

    if verbose:
        print(
            f"Example {config.example_id}: {config.name}\n"
            f"method={method}, states={config.N}, M_max={config.M_max}, "
            f"N_max={config.N_max}, lambda={config.lam}\n"
            f"stationary distribution="
            f"{np.array2string(pi, precision=8)}\n"
            f"sum of reconstructed PMF={result.probability_sum:.12g}\n"
            f"elapsed time={elapsed:.4f} seconds"
        )

    return result


def print_values(
    result: ExampleResult,
    *,
    show_moments: bool = False,
) -> None:
    """Print binomial moments and the PMF."""
    if show_moments:
        print("Binomial moments")
        print("----------------")

        for i, value in enumerate(result.binomial_moments):
            print(f"moment[{i:3d}] = {float(value):.6e}")

        print()

    print("Probability mass function")
    print("-------------------------")

    for i, value in enumerate(result.pmf):
        print(f"P(N = {i:3d}) = {float(value):.6e}")


def save_result(
    result: ExampleResult,
    output_dir: str | Path = ".",
) -> tuple[Path, Path]:
    """Save the PMF and moments as NumPy arrays."""
    output_dir = Path(output_dir)

    output_dir.mkdir(parents=True, exist_ok=True)

    pmf_path = output_dir / f"PMF{result.config.output_prefix}.npy"
    moments_path = output_dir / f"BM{result.config.output_prefix}.npy"

    np.save(pmf_path, result.pmf)
    np.save(moments_path, result.binomial_moments)

    return pmf_path, moments_path


def plot_pmf(
    result: ExampleResult,
    *,
    title: Optional[str] = None,
    figsize: tuple[float, float] = (8, 8),
    dpi: int = 100,
    save_path: Optional[str | Path] = None,
) -> None:
    """Plot the steady-state PMF."""
    fig, ax = plt.subplots(figsize=figsize, dpi=dpi)

    x_values = np.arange(len(result.pmf))

    markerline, stemlines, _ = ax.stem(
        x_values,
        result.pmf,
        linefmt="--",
        markerfmt="o",
        basefmt="",
        label="Theoretical Result",
    )

    markerline.set_markersize(7)
    stemlines.set_linewidth(1.5)

    ax.set_title(
        title or f"Example {result.config.example_id}: Steady-State PMF",
        fontsize=18,
        fontweight="bold",
    )
    ax.set_xlabel(
        "Number of mRNA Molecules",
        fontsize=14,
        fontweight="bold",
    )
    ax.set_ylabel(
        "Probability",
        fontsize=14,
        fontweight="bold",
    )

    ax.tick_params(
        direction="out",
        width=1.5,
        length=6,
        labelsize=12,
    )
    ax.grid(alpha=0.25)
    ax.legend()

    fig.tight_layout()

    if save_path is not None:
        save_path = Path(save_path)
        save_path.parent.mkdir(parents=True, exist_ok=True)

        fig.savefig(
            save_path,
            dpi=300,
            bbox_inches="tight",
        )

    plt.show()


def run_all_examples(
    examples: Dict[int, ExampleConfig],
    *,
    method: Literal["direct", "recursive"] = "direct",
    save: bool = False,
    output_dir: str | Path = "results",
    verbose: bool = True,
) -> Dict[int, ExampleResult]:
    """Run all configurations with the selected method."""
    results: Dict[int, ExampleResult] = {}

    for example_id in sorted(examples):
        result = run_example(
            examples[example_id],
            method=method,
            verbose=verbose,
        )

        results[example_id] = result

        if save:
            save_result(result, output_dir)

    return results


def summarize_results(
    results: Dict[int, ExampleResult],
) -> None:
    """Print one summary row per result."""
    header = (
        f"{'ID':>2}  "
        f"{'Name':<32} "
        f"{'Method':<9} "
        f"{'N':>3} "
        f"{'M_max':>6} "
        f"{'N_max':>6} "
        f"{'PMF sum':>15} "
        f"{'Seconds':>10}"
    )

    print(header)
    print("-" * len(header))

    for example_id in sorted(results):
        result = results[example_id]
        config = result.config

        print(
            f"{example_id:>2}  "
            f"{config.name:<32.32} "
            f"{result.method:<9} "
            f"{config.N:>3} "
            f"{config.M_max:>6} "
            f"{config.N_max:>6} "
            f"{result.probability_sum:>15.10f} "
            f"{result.elapsed_seconds:>10.4f}"
        )
```
